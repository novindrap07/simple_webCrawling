```python
import requests
from bs4 import BeautifulSoup 
import csv 
import pandas as pd
```

```python
url = "https://www.bi.go.id/id/umkm/klaster/produk-ekspor/SentraDetail.aspx?Sentra_ID="
df = []
for page in range(1,104):
    page_url = url+str(page)
    r = requests.get(page_url)
    soup = BeautifulSoup(r.content, "html.parser")
    
    # soup.find_all
    td_al_left = soup.find_all('td', {'align': 'left', 'valign': None, 'nowrap': None})
    quote = {}
    
    ## Data Perusahaan
    quote['Nama_Perusahaan'] = td_al_left[0].text
    quote['Nama_Pemilik'] = td_al_left[2].text 
    quote['Alamat_Perusahaan'] = td_al_left[4].text  
    quote['Provinsi'] = td_al_left[6].text   
    quote['No_Telp/Fax'] = td_al_left[8].text
    quote['Email'] = td_al_left[10].text
    quote['Thn_Berdiri'] = td_al_left[12].text
    quote['Sektor'] = td_al_left[14].text
    quote['Sub_Sektor'] = td_al_left[16].text
    
    ## Aspek Produksi
    quote['Jenis_Produk'] = td_al_left[19].text
    quote['Sps_Produk_1'] = td_al_left[23].text
    quote['Sps_Produk_2'] = td_al_left[25].text
    quote['Sps_Produk_3'] = td_al_left[27].text
    quote['Kualitas_Produk_Grd_A'] = td_al_left[31].text
    quote['Kualitas_Produk_Grd_B'] = td_al_left[33].text
    quote['Kualitas_Produk_Grd_C'] = td_al_left[35].text
    quote['Merk_Produk'] = td_al_left[37].text
    quote['Produk_Stdr/Psn'] = td_al_left[39].text
    quote['Kemampuan_Produksi/Bulan'] = td_al_left[41].text
    
    ## Aspek Pasar & Distribusi
    quote['Tujuan_Lokal'] = td_al_left[46].text
    quote['Tujuan_Nasional'] = td_al_left[48].text
    quote['Tujuan_Ekspor'] = td_al_left[50].text
    quote['Pengalaman_Ekspor'] = td_al_left[52].text
    quote['Mekanisme_Ekspor'] = td_al_left[54].text
    quote['Tahun_Mulai_Ekspor'] = td_al_left[56].text
    
    ## Sumber Pembiayaab
    quote['Bank'] = td_al_left[59].text
    quote['Lembaga_Keuangan_Lain'] = td_al_left[61].text
    
    # Append it!
    df.append(quote) 
    
# filename = 'Data_UMKM_Produk_Ekspor.csv'
# with open(filename, 'ww') as f: 
#     w = csv.DictWriter(f,['Nama_Perusahaan','Nama_Pemilik','Alamat_Perusahaan','Provinsi','No_Telp/Fax',
#                          'Email','Thn_Berdiri','Sektor','Sub_Sektor','Jenis_Produk','Sps_Produk_1',
#                          'Sps_Produk_2','Sps_Produk_3','Kualitas_Produk_Grd_A','Kualitas_Produk_Grd_B',
#                          'Kualitas_Produk_Grd_C','Merk_Produk','Produk_Stdr/Psn','Kemampuan_Produksi/Bulan',
#                          'Tujuan_Lokal','Tujuan_Nasional','Tujuan_Ekspor','Pengalaman_Ekspor','Mekanisme_Ekspor',
#                          'Tahun_Mulai_Ekspor','Bank','Lembaga_Keuangan_Lain']) 
#     w.writeheader() 
#     for quote in quotes: 
#         w.writerow(quote) 
```
